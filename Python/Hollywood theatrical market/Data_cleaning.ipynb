## User define funtions 

few funtions to avoid repatative tasks 

```python
def rmv_character(clm_name,df):
    """ Remove non-numaric values from column of given df
    clm_name : columns from which non-numaric values are to be removed
    df : name of dataframe from which non-numaric values are to be removed
    """
    for clm in clm_name:
        df[clm] = df[clm].apply(lambda x : re.sub('[^.0-9]','',x))
```
